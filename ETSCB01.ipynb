````python
#Dual y axis plot
ax=df.plot(x='Molecular Weight',y='ESOL predicted log solubility in mols per litre', kind='scatter',color='pink')
df.plot(x='Molecular Weight',y='measured log solubility in mols per litre', ax=ax, kind='scatter',color='purple')
plt.title('Multiple y plots')
plt.xlabel('Molecular Weight')
plt.ylabel('Solubility')
plt.legend()
plt.grid(True)

x=df['Molecular Weight']
y1=['ESOL predicted log solubility in mols per litre']
y2=['measured log solubility in mols per litre']
m1, b1= np.polyfit(x, y1, 1)
m2, b2= np.polyfit(x, y2, 1)

plt.plot(x,m1*x+b1,color='magenta',label='ESOL predicted log solubility in mols per litre')
plt.plot(x,m2*x+b2,color='cyan',label='measured log solubility in mols per litre')
````
